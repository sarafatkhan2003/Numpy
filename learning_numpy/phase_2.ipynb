```python
a = np.array([2,4,5,6,3])

print(a > 2)
print(a < 4)
print(a >= 4)
print(a >= 6)



```

***Broadcasting***

*What is Broadcasting?*

 - *Broadcasting allows NumPy to perform operations on arrays of different shapes.*

```python
arr = np.array([7,5,2])
print(arr + 5)

print(arr - 2)
print(arr * 2)
print(arr / 3 )
```

```python
mat = np.array([[1,2,3],
           [4,5,6], 
           [5,4,3]])

print(mat + 2)
print(mat - 3)
print(mat * 4)
print(mat / 2)
```

***Broadcasting Rules***

 
*Broadcasting works if from right to left:*

- Dimensions are equal

- One of them is 1


```python
a = np.array([[1,2,3],
             [2,1,3],
             [7,6,1]])

b = np.array([7,6,5])
print(a+b)
```

```python
a = np.array([[1,2,3],
              [4,5,6]])   # (2,3)

b = np.array([10,20,30])  # (3,)

print(a + b)
```

**Arithmetic + Comparison Together**

```python
a = np.array([10, 20, 30])

result1 = (a * 2) + 5
result2 = (a / 5) * 2
print(result1)
print (result2)
```

**ðŸŽ¯ Interview Answer**

**NumPy avoids Python loops because vectorized operations execute in optimized C code, making them faster and memory-efficient.**

```python

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)
print(a * 10)
print(a > 2)
print(a < 3)

matrix = np.array([[1,2,3],
                   [4,5,6]])

print(matrix + 100)
```
