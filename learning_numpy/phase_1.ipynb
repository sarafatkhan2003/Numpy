**Why we use NumPy?**
- NumPy is faster than Python lists because it stores data in contiguous memory and uses vectorized operations implemented in C.

*Problem with Python Lists*

*Python lists are:*

- Slow for math operations

- Not memory efficient

- Not optimized for numerical computing

*NumPy provides:*

- Fast arrays

- Vectorized operations

- Written in C (speed ðŸš€)

````python
# Python Example
a = [1, 2, 3]
b = [4, 5, 6]

c = [a[i] + b[i] for i in range(len(a))]
print(c)
````

````python
# NUmpy eaxmple
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

c = a + b
print(c)
````

***Indexing & Slicing***

````python
# 1D Indexing
arr = np.array([1,2,3,4,5,6,7])
print(arr[0])
print(arr[-1])
print(arr[-4])
print(arr[4])
````

````python
# 2D Indexing [1,3,4,5],[1,4,3,5]
print(arr[0, 3])
print(arr[1, 1]) 
````

````python
# Slicing
arr = np.array([2,3,4,5,6,7,8,9,10])
print(arr[1:4])
print(arr[:4])
print(arr[::-2])
````
